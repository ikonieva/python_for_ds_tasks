1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.

```python
def learning(x):
  return 10*x - 5
```

```python
import matplotlib.pyplot as plt
import numpy as np

hours = np.array(range(0,11))
plt.plot(hours, learning(hours), 'o')
plt.xlabel("Години навчання")
plt.ylabel("Бали")
plt.yticks(learning(hours))
plt.xticks(hours);
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

x = sp.symbols('x')

f_sin = sp.sin(x)
f_ln = sp.ln(x)

derivative_f_sin = sp.diff(f_sin, x)
derivative_f_ln = sp.diff(f_ln, x)

print("Похідна функції f(x)=sin(x):", derivative_f_sin)
print("Похідна функції f(x)=ln(x):", derivative_f_ln)
```

```python
def f_sin(x):
  return np.sin(x)

def f_ln(x):
  return np.log(x)

def df_sin(x):
  return np.cos(x)

def df_ln(x):
  return 1/x

xs = np.linspace(0, 10, 100)
np.random.seed(42)
x_point = np.random.choice(xs, 1, replace=False)
```

```python
plt.plot(xs, f_sin(xs), label='sin(x)', color='green')
plt.plot(xs, df_sin(xs), label='cos(x)', color='orange')

f_point = f_sin(x_point)
df_point = df_sin(x_point)
tangent_line = df_point * (xs - x_point) + f_point

plt.scatter(x_point, f_point, color='red')
plt.plot(xs, tangent_line, '--', color='red', label=f'Tangent at x = {x_point}')

plt.title("Графік функції f(x)=sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.plot(xs, f_ln(xs), label='ln(x)', color='green')
plt.plot(xs, df_ln(xs), label='1/x', color='orange')

f_point = f_ln(x_point)
df_point = df_ln(x_point)
tangent_line = df_point * (xs - x_point) + f_point

plt.scatter(x_point, f_point, color='red')
plt.plot(xs, tangent_line, '--', color='red', label=f'Tangent at x = {x_point}')

plt.title("Графік функції f(x)=ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
f = x**2 - 5 * x + 1
df = sp.diff(f, x)
print("Функція витрат компанії:", f)
print("Похідна функції витрат компанії:", df)
```

```python
def f_spend(x):
  return x**2 - 5 * x + 1

def df_spend(x):
  return 2*x - 5

products = np.array(range(0,100))

plt.plot(products, f_spend(products), label='x**2 - 5 * x + 1', color='green')
plt.plot(products, df_spend(products), label='2*x - 5', color='orange')
plt.title("Графік функції f(x)=x**2 - 5 * x + 1 та її похідної")
plt.xlabel("products")
plt.ylabel("spends")
plt.legend()
plt.grid(True)

plt.show();
```
